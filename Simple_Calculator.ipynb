```python
import ipywidgets as widgets  # Library for creating interactive UI elements in Jupyter/Colab
from IPython.display import display  # Used to display widgets in the output cell

def on_button_click(symbol):
    """
    Handles button clicks to update the calculator display.
    - Evaluates expressions when '=' is pressed.
    - Clears input when 'C' is pressed.
    - Appends clicked symbol to the display otherwise.
    """
    if symbol == '=':
        try:
            result = str(eval(entry.value))  # Evaluate the mathematical expression
            entry.value = result  # Display the result
        except Exception:
            entry.value = "Error"  # Handle invalid expressions
    elif symbol == 'C':
        entry.value = ""  # Clear the input field
    else:
        entry.value += symbol  # Append the pressed button value to the display

# Create the entry field for displaying input/output
entry = widgets.Text(value='', layout=widgets.Layout(width='100%', height='40px'))

# Create a grid layout for calculator buttons
grid_layout = widgets.GridBox(layout=widgets.Layout(grid_template_columns="repeat(4, 1fr)", gap="5px"))

# Define button labels in a 4x4 grid format
buttons = [
    ('7', '8', '9', '/'),
    ('4', '5', '6', '*'),
    ('1', '2', '3', '-'),
    ('C', '0', '=', '+')
]

# Create button widgets and assign actions
button_widgets = []
for row in buttons:
    for char in row:
        button = widgets.Button(description=char, layout=widgets.Layout(width='50px', height='50px'))
        button.on_click(lambda b, ch=char: on_button_click(ch))  # Assign click event to button
        button_widgets.append(button)

# Add buttons to the grid layout
grid_layout.children = button_widgets

# Display the calculator UI
display(entry, grid_layout)
```
